```python
import pandas as pd
csv_test = pd.read_csv('notExercise.csv', encoding='cp949', sep=",")
csv_test
```

```python
mf = csv_test.loc[1:2, '분류':'기타']
age = csv_test.loc[3:8,'분류':'기타']
school = csv_test.loc[9:12,'분류':'기타']
```

```python
header = mf.columns.tolist()
header = header[1:5]
header
```

```python
gender = mf.loc[1:2, '분류'].tolist()
born = age.loc[3:8, '분류'].tolist()
study = school.loc[9:12, '분류'].tolist()

data = []
data1 = []
data2 = []
for i in range(len(header)):
    data.append(mf.loc[1:2, header[i]].tolist())
    data1.append(age.loc[3:8, header[i]].tolist())
    data2.append(school.loc[9:12, header[i]].tolist())
```

### 성별 운동하지 않는 이유 비율
---

```python
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(8, 8))

for i in range(len(header)):
    plt.subplot(2, 2, i+1)
    plt.title(header[i])
    plt.pie(data[i], labels=gender, autopct='%1.1f%%')
plt.show()
```

### 연령별 운동하지 않는 이유 비율
---

```python
plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(12, 8))

for i in range(len(header)):
    plt.subplot(2, 2, i+1)
    plt.title(header[i])
    plt.pie(data1[i], labels=born, autopct='%1.1f%%')
plt.show()
```

### 학년별 운동하지 않는 이유 비율
---

```python
plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(12, 8))

for i in range(len(header)):
    plt.subplot(2, 2, i+1)
    plt.title(header[i])
    plt.pie(data2[i], labels=study, autopct='%1.1f%%')
plt.show()
```
